```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Load dataset
data = pd.read_csv('wbc.csv')
data.head()
```

```python
data = data.drop(columns=['id'])
data.head()
```

```python
print(data.isnull().sum())
```

```python
data = data.drop(columns=['Unnamed: 32'])
```

```python
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
data_numerik =  data.drop(columns=['diagnosis']).select_dtypes(include=['float64', 'int64']).columns

std = StandardScaler()
data[data_numerik] = std.fit_transform(data[data_numerik])
data.head()
```

```python
X = data.drop(columns=['diagnosis'])  # Fitur
y = data['diagnosis']  # Target

# Lakukan stratified split untuk data latih dan data uji dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)

# Cek masing-masing ukuran data
print(f'Jumlah label data asli:\n{y.value_counts()}')
print(f'Jumlah label data train:\n{y_train.value_counts()}')
print(f'Jumlah label data test:\n{y_test.value_counts()}')

# Print shapes of the datasets
print(f'Training data shape: {X_train.shape}')
print(f'Test data shape: {X_test.shape}')
print(f'Training target shape: {y_train.shape}')
print(f'Test target shape: {y_test.shape}')
```
